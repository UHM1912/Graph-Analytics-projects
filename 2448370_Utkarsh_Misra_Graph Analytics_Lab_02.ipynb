```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph(graph, title, edge_labels=None):
    plt.figure(figsize=(7, 6))
    pos = nx.spring_layout(graph, seed=42)
    nx.draw(graph, pos, with_labels=True, node_size=500, node_color='orange',
            font_size=10, font_weight='bold', edge_color='gray')
    if edge_labels:
        nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
    plt.title(title)
    plt.show()
```

# 1. Directed Graph

```python
directed_graph = nx.DiGraph([(0, 1), (1, 2), (2, 0)])
draw_graph(directed_graph, "Directed Graph")
```

# 2. Weighted Graph

```python
weighted_graph = nx.Graph()
weighted_graph.add_edge(0, 1, weight=4.5)
weighted_graph.add_edge(1, 2, weight=2.1)
weighted_graph.add_edge(2, 0, weight=3.3)
edge_labels = nx.get_edge_attributes(weighted_graph, 'weight')
draw_graph(weighted_graph, "Weighted Graph", edge_labels)
```

# 3. Bipartite Graph

```python
B = nx.Graph()
B.add_nodes_from([1, 2, 3, 4], bipartite=0)
B.add_nodes_from(['a', 'b', 'c'], bipartite=1)
B.add_edges_from([(1, 'a'), (2, 'b'), (3, 'c'), (4, 'a')])
plt.figure(figsize=(7, 6))
nx.draw(B, with_labels=True, node_size=500, node_color='lightcoral',
        font_size=10, font_weight='bold', edge_color='gray')
plt.title("Bipartite Graph")
plt.show()
```

# 4. Undirected Graph

```python
undirected_graph = nx.Graph([(0, 1), (1, 2), (2, 0)])
draw_graph(undirected_graph, "Undirected Graph")
```

# 5. Tree Graph

```python
tree_graph = nx.balanced_tree(r=2, h=3)
draw_graph(tree_graph, "Tree Graph")
```

# 6. Pseudograph (Graph with Self-loops)

```python
pseudograph = nx.MultiDiGraph([(0, 1), (1, 2), (2, 0), (2, 2)])
draw_graph(pseudograph, "Pseudograph (Graph with Self-loops)")
```


# 7. Scale-free Graph

```python
scale_free_graph = nx.scale_free_graph(10)
draw_graph(scale_free_graph, "Scale-free Graph")
```

# 8. Complete Graph

```python
complete_graph = nx.complete_graph(5)
draw_graph(complete_graph, "Complete Graph")
```

# 13. Grid Graph (2D lattice)

```python
grid_graph = nx.grid_2d_graph(3, 3)
plt.figure(figsize=(6, 6))
nx.draw(grid_graph, with_labels=True, node_size=500, node_color='orange',
        font_size=10, font_weight='bold')
plt.title("Grid Graph (3x3)")
plt.show()
```

# 14. Random Graph (Erdős–Rényi)

```python
random_graph = nx.erdos_renyi_graph(10, 0.3)
draw_graph(random_graph, "Random Graph (Erdős–Rényi)")
```

# 15. Small-World Graph (Watts–Strogatz)

```python
small_world = nx.watts_strogatz_graph(10, 4, 0.3)
draw_graph(small_world, "Small-World Graph (Watts–Strogatz)")
```

# 16. Hypercube Graph

# 17. Barbell Graph

```python
barbell_graph = nx.barbell_graph(4, 2)
draw_graph(barbell_graph, "Barbell Graph")
```


# 18. Lollipop Graph

```python
lollipop_graph = nx.lollipop_graph(4, 3)
draw_graph(lollipop_graph, "Lollipop Graph")
```
